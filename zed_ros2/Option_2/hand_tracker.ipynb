This file will use the EXTRACTED images and point cloud from the ros2bag file for Handtracking.
It also relies on a customized package called "tracking.py"

```python
import PyQt5
%matplotlib qt
from tracking import handtracking
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys
import datetime
import glob
```

```python
detector = handtracking()
```

```python
fig = plt.figure()
plt.ion()
ax = fig.add_subplot(111, projection='3d')
```

```python
filenames = glob.glob('left_images/*.jpg')
filenames = sorted(filenames)
```

```python
filenames
```

```python
for image_path in filenames:
    print(image_path)
    image = cv2.imread(image_path)
    image = detector.detect_landmarks(image)
    cv2.imshow("test", image)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows()
```
